```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named dataFrame with a column 'Brand'

# Count occurrences of each brand
brand_counts = dataFrame['Brand'].value_counts()

# Calculate the percentage of each brand
brand_percentages = brand_counts / brand_counts.sum() * 100

# Filter brands that represent less than 1% of the total
filtered_brands = brand_percentages[brand_percentages >= 1]

# Get the sum of counts for brands representing less than 1%
other_count = brand_counts[brand_percentages < 2.4].sum()

# If there are brands less than 1%, combine them into 'Other'
if other_count < 0:
    filtered_brands['Other'] = other_count

# Plotting the pie chart with percentages
plt.figure(figsize=(8, 10))
plt.pie(filtered_brands, labels=filtered_brands.index, autopct='%1.1f%%')
plt.title('Percentage of Electric Vehicle Brands')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the occurrences of each region directly from the DataFrame
ev_counts = dataFrame['County'].value_counts()

# Extract region names and their counts
regions = ev_counts.index
counts = ev_counts.values

# Set the figure size (width, height)
plt.figure(figsize=(60, 30))  # Adjust width and height according to your preference

# Define a color map
colors = plt.cm.tab10(range(len(regions)))

# Plotting each bar separately to create legend labels
for i in range(len(regions)):
    plt.bar(regions[i], counts[i], color=colors[i], label=regions[i])

# Adding title and labels
plt.title('Geographical Distribution of Electric Vehicles')
plt.xlabel('Regions')
plt.ylabel('Number of Electric Vehicles')

plt.ylim(0, max(counts) * 0.5)
# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Adding legend
plt.legend(prop={'size': 50}, loc='upper left', bbox_to_anchor=(1, 1))

# Displaying the bar chart

plt.show()
```

```python
plt.figure(figsize=(5,10), dpi=100)

plt.style.use('default')

tesla = dataFrame.loc[dataFrame.Brand == "TESLA"]['Electric Range']
nissan = dataFrame.loc[dataFrame.Brand == "NISSAN"]['Electric Range']
chev = dataFrame.loc[dataFrame.Brand == "CHEVROLET"]['Electric Range']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([tesla, nissan, chev], labels=['TESLA','NISSAN','CHEVROLET'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('EV Car Brands Comparisson')
plt.ylabel('Electric Range (km)')

for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')

plt.show()
```

```python
selectedDataFrame = dataFrame[dataFrame['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] == 'Clean Alternative Fuel Vehicle Eligible']
selectedArray = selectedDataFrame['Brand'].unique()

def find_common_elements(array1, array2):
    common_elements = []
    for element in array1:
        if element in array2:
            common_elements.append(element)
    return common_elements

# Capitalizing the elements of selectedArray
capitalized_array = [element.capitalize() for element in selectedArray]

# Example array
az_cars = ['Audi', 'Volkswagen', 'Toyota', 'Honda', 'BYD', 'Tesla', 'Zeekr', 'Polestar', 'BMW', 'Mercedes-Benz', 'LJ Auto', 'NIO', 'Changan', 'Haval', 'Skywell']

common_elements = find_common_elements(capitalized_array, az_cars)
print("Common car brands:", common_elements)
```
